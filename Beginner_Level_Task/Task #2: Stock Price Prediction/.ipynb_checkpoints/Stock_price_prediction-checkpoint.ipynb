```python
training_set = df['Open']
training_set = pd.DataFrame(training_set)
training_set = training_set.reset_index()['Open']
training_set
```

```python
training_set.shape
```

```python
## Feature Scaling

from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0,1))
training_set_scaled = sc.fit_transform(np.array(training_set).reshape(-1,1))
```

```python
train_size = int(len(training_set_scaled))*0.65
test_size = int(len(training_set_scaled)) - train_size
```

```python
df_train = training_set_scaled[0:int(train_size)]
# df_test = training_set_scaled[-int(test_size):]
df_test = training_set_scaled[int(train_size):len(training_set_scaled),:]
```

```python
def select_dataset_timestep(dfl,timestep=1):
    df_X = []
    df_Y = []
    for i in range(len(dfl)-timestep - 1):
        df_X.append(dfl[i:(i+timestep),0])
        df_Y.append(dfl[(i+timestep),0])
    
    return np.array(df_X), np.array(df_Y)
```

```python
df_train.shape
```

```python
timestamp = 100

X_train, Y_train = select_dataset_timestep(df_train,timestamp)
X_test, Y_test = select_dataset_timestep(df_test,timestamp)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
Y_test.shape
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1],1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1],1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
import tensorflow as tf
from sklearn.metrics import mean_squared_error

model = tf.keras.models.Sequential([
        tf.keras.layers.LSTM(60,return_sequences=True,input_shape=(100,1)),
#         tf.keras.layers.Dropout(0.2),
        tf.keras.layers.LSTM(60,return_sequences=True),
#         tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(1)
])

model.compile(loss='mean_squared_error', optimizer='adam', metrics='acc')
hist = model.fit(X_train, Y_train, validation_data = (X_test, Y_test), epochs = 30, batch_size = 64, verbose = 1)

```

```python
model.summary()
```

```python
train_predict =model.predict(X_train)
test_predict =  model.predict(X_test)
#Transformback to original form

train_predict = train_predict.reshape(train_predict.shape[0],train_predict.shape[1])
test_predict = test_predict.reshape(test_predict.shape[0],test_predict.shape[1])
train_predict=sc.inverse_transform(train_predict)
test_predict=sc.inverse_transform(test_predict)
```

```python
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(Y_train, train_predict))
```
